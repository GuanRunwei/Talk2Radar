```python
import os
import shutil

# 
folder_path = 'pred_instances_3d'

# 
new_folder_path = 'converted_instances_3d'

# 
if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)

# txt
new_names_file = '../data/vod/radar_5frames/ImageSets/val.txt'
# new_names_file = '../data/vod/radar_5frames/ImageSets/test.txt'

# 
with open(new_names_file, 'r') as file:
    new_names = file.read().splitlines()

# 
assert len(new_names) == len(os.listdir(folder_path)), "The number of new names does not match the number of files."

# 
for i, filename in enumerate(sorted(os.listdir(folder_path)), start=1):
    old_path = os.path.join(folder_path, filename)
    new_filename = new_names[i-1] + os.path.splitext(filename)[1]  # 
    new_path = os.path.join(new_folder_path, new_filename)
    shutil.move(old_path, new_path)
    print(f'Moved {old_path} to {new_path}')

print("All files have been moved to the new location successfully.")
```
